```python
dfN=df[['coaster_name','Location', 'Status', 
      'Manufacturer', 
       'year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean',
       'speed_mph', 'height_ft',
       'Inversions_clean', 'Gforce_clean']]
      
```

```python
dfN
```

```python
dfN.dtypes
```

```python
dfN.opening_date_clean=pd.to_datetime(dfN.opening_date_clean)
```

```python
dfN.dtypes
```

```python
dfN.isna().sum()
```

```python
df2=dfN.loc[~dfN.duplicated(subset=['coaster_name','Location','opening_date_clean'])].reset_index(drop=True)
```

```python
df2.shape
```

```python
df2['year_introduced'].value_counts().head(10).plot(kind='bar',title='Top Years Coasters have introduced')
```

```python
df2['speed_mph'].plot(kind='kde',title='Speed (mph)')
```

```python
sns.scatterplot(x='speed_mph',y='height_ft',data=df2,hue='year_introduced')
```

```python
corrr=df2[['year_introduced','speed_mph','height_ft','Inversions_clean','Gforce_clean']].dropna().corr()
sns.heatmap(corrr,annot=True)
```

```python
df2['speed_mph'].idxmax()
```

```python
df2.groupby('Location')['speed_mph'].agg(['mean','count']).query('count >= 10').sort_values('mean')['mean'].plot(kind='bar',title='average coaster speed per location')
```
